#### Movie Gross data:

```python
!ls
```

```python
# IMPORT MOVIE GROSS DATA
df_moive_gross = pd.read_csv('bom.movie_gross.csv.gz')
```

```python
df_moive_gross.info()
```

```python
df_moive_gross
```

```python
df_moive_gross['domestic_gross'] = df_moive_gross.apply(lambda x: "{:,}".format(x['domestic_gross']), axis=1)
df_moive_gross
```
